## joins

```python
df_1 = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7,8,9]], columns =['A', 'B','C'])
```

```python
df_2 = pd.DataFrame([[1,'b'], [4, 'd']], columns = ['A', 'D'])
```

```python
df_1
```

```python
df_2
```

```python
pd.merge(df_1, df_2, on='A')
```

```python
df_2 = pd.DataFrame([[1,'joe', 'doe'], 
                     [4, 'jane', 'smith'],
                     [5, 'jon', 'johnson']
                    ], 
                    columns = ['id', 'firstname', 'lastname'])
```

```python
df_2
```

```python
df_1 = pd.DataFrame([[1, 'A', 'C'], 
              [4, 'D', 'A'], 
              [7, 'F', 'F'],
              [1, 'C', 'D']], 
             columns =['id', 'math', 'biology'])
```

```python
df_1
```

```python
df_2
```

## inner join

```python
pd.merge(df_1, df_2, on='id', how = 'inner')
```

## right join

```python
pd.merge(df_1, df_2, on='id', how = 'right')
```

## left join

```python
pd.merge(df_1, df_2, on='id', how='left')
```

## outer join

```python
df_3 = pd.merge(df_1, df_2, on='id', how = 'outer')
```

## pivots, spreads, gather, melts

```python
df_1 = pd.DataFrame(
    {
        'name': ['John', 'Jane', 'Mary', 'John', 'Jane', 'Mary'],
        'groupvar': ['A', 'A', 'A', 'B', 'B', 'B'],
        'measure': [2, 11, 1, 3, 4, 5]
    }
)
```

```python
df_1
```

```python
df_1.pivot(index = 'name', columns = 'groupvar', values = 'measure')
```

```python
df_2 = pd.DataFrame(
    {
        'name': ['A', 'B', 'C'],
        'group_A': [1, 2, 3],
        'group_B': [4, 5, 6]
    }
)
```

```python
df_2
```

```python
df_2.melt(id_vars = ['name'])
```
