```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
weight_df = pd.read_csv('weight.csv')
```

```python
weight_df
```

```python
sns.scatterplot(x = 'height', y = 'weight', data = weight_df)
```

```python
%rep 12
```

```python
sns.scatterplot(x = 'height', y = 'weight', hue = 'gender', data = weight_df)
```

```python
sns.relplot(x = 'height', 
            y = 'weight', 
            hue = 'gender', 
            col = 'race', 
            col_wrap = 3, 
            data = weight_df)
```

```python
sns.lmplot(x = 'height',
           y = 'weight',
           hue = 'gender',
           col = 'race',
           col_wrap = 3,
           scatter_kws = {'s': 10, 'alpha': 0.1},
           ci = None,
           data = weight_df)
```

```python
sns.catplot(x = 'gender', y = 'weight', ci = 'sd', kind = 'bar', data = weight_df)
```
