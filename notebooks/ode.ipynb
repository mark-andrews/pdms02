```python
import numpy as np
from scipy.integrate import odeint

import matplotlib.pyplot as plt
```

# A chaotic attractor

$$
\begin{aligned}
\dot{x} &= -y -z,\\
\dot{y} &= x + ay,\\
\dot{z} &= b + z(x -c)
\end{aligned}
$$

```python
def roessler(X, t, a, b, c):
    
    x, y, z = X
    
    dx = -y - z
    dy = x + a*y
    dz = b + z*(x -c)
    
    return dx, dy, dz
```

```python
a, b, c = 0.2, 0.2, 5.7
```

```python
t = np.linspace(0, 100, 10000)
```

```python
trajectory = odeint(roessler, (0,0,0), t, args=(a, b, c))
```
