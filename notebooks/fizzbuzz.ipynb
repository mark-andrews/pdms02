# The fizz-buzz test

Take a sequence of numbers. If the number is divisble by 3, the print "fizz". If it divisible by 5, then print "buzz". If divisible by both 3 and 5, the print "fizzbuzz". Otherwise, print the number itself.

For starters, use the integers 1 to 99.


```python
def divs(x, k):
    return x % k == 0
```

```python
for x in range(1, 100):
    
    if divs(x, 3) and divs(x, 5):
        print('fizzbuzz')
    elif divs(x, 3):
        print('fizz')
    elif divs(x, 5):
        print('buzz')
    else:
        print(x)
```

```python
def fizzbuzz(n, kf=3, kb=5, fizz='fizz', buzz='buzz'):
    for x in range(1, n):

        if divs(x, kf) and divs(x, kb):
            print(fizz + buzz)
        elif divs(x, kf):
            print(fizz)
        elif divs(x, kb):
            print(buzz)
        else:
            print(x)
```

```python
fizzbuzz(100, kf=7, kb=11, fizz = 'foo', buzz = 'bar')
```
