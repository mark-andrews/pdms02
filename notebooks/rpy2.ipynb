```python
import rpy2
import pandas as pd
```

```python
from rpy2 import robjects as ro
from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri

pandas2ri.activate()
R = ro.r
base = importr('base')
```

```python
lme4 = importr('lme4')
```

```python
weight_df = pd.read_csv('weight.csv')

m = R.lm('weight ~ height', data = weight_df)

print(base.summary(m))
```

```python
R.coef(m)
```

```python
print(base.summary(m).rx2('coefficients'))
```
